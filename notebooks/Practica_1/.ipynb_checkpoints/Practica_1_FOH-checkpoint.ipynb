## Introducción

La simulación de robots en entornos virtuales se ha convertido en una herramienta esencial en el campo de la robótica. Permite a los ingenieros y desarrolladores modelar y analizar el comportamiento de robots sin necesidad de construir físicamente el hardware. Esto no solo ahorra tiempo y recursos, sino que también facilita la identificación y corrección de errores en las fases tempranas del diseño.

En esta práctica, utilizamos RViz, que es un visualizador de robots dentro del marco de trabajo ROS (Robot Operating System). RViz nos permite ver y manipular representaciones gráficas de robots y sus componentes en un entorno tridimensional. A través de este visualizador, podemos observar cómo se comporta un sistema robótico en diferentes configuraciones y condiciones.

## Previo
1. Definición de URDF.

El formato URDF (Unified Robot Description Format) es una especificación basada en XML que permite describir detalladamente un robot, incluyendo su aspecto visual, cinemática, dinámica y modelo de colisión. A través de elementos XML, se definen características como la forma, tamaño, colores, enlaces, articulaciones, propiedades físicas (masa, inercia) y las partes del robot que pueden colisionar. Con un modelo URDF, es posible realizar simulaciones en entornos como Gazebo o RVIZ2, facilitando el análisis y control del comportamiento del robot [[1]](#1).

2. Tipos de juntas que están definidas dentro del formato URDF.

En un robot, cada una de las partes rígidas que lo componen se denomina eslabón o link, y la unión entre dos eslabones se llama articulación o joint. Dependiendo del tipo de movimiento que permite la articulación, en URDF existen diferentes tipos:

- revolute: permite el giro sobre un eje, con un rango limitado especificado por límites superior e inferior.
- continuous: permite el giro alrededor de un eje, pero sin límites de rotación.
- prismatic: permite el deslizamiento a lo largo de un eje, con un rango limitado por límites superior e inferior.
- fixed: no permite ningún movimiento, pero es útil para asociar un sistema de referencia fijo, por ejemplo, para sensores o cámaras.
- floating: permite movimiento en los 6 grados de libertad (traslación y rotación en los tres ejes).
- planar: permite movimiento en un plano perpendicular a un eje [[1]](#1).

3. Definición de elementos "primitivos" que se pueden utilizar para definir un robot en ROS.

- Links (eslabones): Representan las partes rígidas del robot, que pueden ser de diferentes formas y tamaños. Se describen utilizando modelos 3D, generalmente en formatos como URDF (Unified Robot Description Format).

- Joints (articulaciones): Son los mecanismos que conectan dos links y permiten el movimiento entre ellos. Pueden ser de varios tipos, como revolute (rotación), prismatic (deslizamiento), fixed (fijo), entre otros.

- Geometry: Este elemento describe la forma y el tamaño de un eslabón o link del robot. Se especifica usando figuras básicas como cajas, cilindros o esferas. La geometría es fundamental para las simulaciones, ya que determina cómo el robot interactuará con su entorno.

- Visual: Este componente define cómo se verá el eslabón del robot en simulaciones o visualizaciones. Incluye la geometría del eslabón y también puede incluir características estéticas como el color y la textura. Esto es importante para la representación gráfica del robot en entornos de simulación como Gazebo o RVIZ.

- Collision: Esta sección describe cómo se comporta el eslabón en términos de colisiones con otros objetos. Al igual que la geometría visual, incluye una descripción de la forma del eslabón, pero está optimizada para cálculos de colisión, lo que significa que puede ser más simple que la geometría visual. Esto ayuda a prevenir errores en simulaciones al detectar interacciones físicas entre el robot y su entorno.

- Inertial (Inercial): Este elemento describe las propiedades físicas del eslabón relacionadas con su masa y cómo se comporta en movimiento. Incluye la masa del eslabón, la ubicación del centro de masa y la matriz de inercia, que indica cómo el eslabón resistirá la rotación. Esto es esencial para simular correctamente el comportamiento dinámico del robot.


4. Que es un nodo y que es un topico en ROS 2
En ROS2, un nodo es un ejecutable que utiliza el sistema ROS para comunicarse con otros nodos. Los nodos son componentes fundamentales en la arquitectura de ROS, ya que permiten la interacción y el intercambio de información entre diferentes partes de un robot o un sistema distribuido.

Un tópico es un canal de comunicación en el que los nodos pueden publicar y suscribirse a mensajes. Los nodos pueden publicar mensajes a un tópico específico o suscribirse a un tópico para recibir mensajes. Los mensajes son tipos de datos utilizados en ROS que permiten el intercambio de información entre nodos a través de estos tópicos. En resumen, los nodos envían y reciben información a través de los tópicos, facilitando la comunicación en un sistema robótico [[3]](#3).

5. Que es RVIZ.
RViz es una herramienta de software para visualización y depuración de robots. Le permite mostrar e interactuar con varios tipos de datos, como imágenes, nubes de puntos, mapas, trayectorias, etc. También puede usar RViz para interactuar con ROS y suscribirse y publicar temas, servicios y acciones de ROS [[4]](#4).



## Desarrollo

1. El código comienza definiendo un robot llamado "basic_link". Se crea un enlace vacío llamado "world" que representa el marco de referencia fijo del mundo. A continuación, se establece una articulación fija (joint) llamada "world_to_base_link" que conecta el enlace "world" con el "base_link", que es la base del robot. Esta articulación fija indica que el "base_link" no se moverá con respecto al mundo.

```js
<robot name="basic_link">

    <link name="world">
    </link>

    <!--Definicion de la junta del link base a la tierra-->
    <joint name="world_to_base_link" type="fixed">
        <parent link="world"/>
        <child link="base_link"/>
        <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
    </joint>
```

2. Se define el enlace "base_link" con sus propiedades visuales, de colisión e inerciales:

Visual: Se especifica un cubo (box) con dimensiones de 0.1 x 0.12 x 0.15 metros. El origen del enlace está desplazado en z por 0.075 metros. Se asigna un material de color gris con valores RGBA (0.5, 0.5, 0.5, 1.0).

Colisión: Se utiliza la misma geometría y origen que en la sección visual para definir las propiedades de colisión.

Inercial: Se establece una masa de 0.5 kg. El centro de masa está en el origen del enlace. La matriz de inercia se define con los valores ixx="0.1", iyy="1.0", izz="1.0", y los productos de inercia son cero.

```js
    <!--Definicion de la Base-->
    <link name="base_link">
        <!--Descripcion de la parte visual del eslabon-->
        <visual>
            <!--Posicion y orientacion de origen del eslabon-->
            <origin xyz="0.0 0.0 0.075" rpy="0.0 0.0 0.0"/>
            <!--Definición de la geometria del eslabón-->
            <geometry>
                <box size="0.1 0.12 0.15"/>             
            </geometry>
            <!--Definicion del color del eslabon-->
            <material name="grey">
                <color rgba="0.5 0.5 0.5 1.0" />
            </material>      
        </visual>
        <!--Descripcion de las relaciones de colision-->
        <collision>
            <!--Posicion y orientacion de origen del eslabon-->
            <origin xyz="0.0 0.0 0.075" rpy="0.0 0.0 0.0"/>
            <!--Definicion de la geometria del eslabon-->
            <geometry>
                <box size="0.1 0.12 0.15"/>             
            </geometry>          
        </collision>
        <!--Descripcion de la inercial del eslabon-->
        <inertial>
            <!--masa del eslabon-->
            <mass value="0.5"/>
            <!--Ubicacion del sistema inercial asociado al centro de masa-->
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
            <!--Matriz de inercia en el centro de masa-->
            <inertia ixx="0.1" ixy="0.0" ixz="0.0" iyy="1.0" iyz="0.0" izz="1.0"/> 
        </inertial>
    </link>
```

3. Se añade un nuevo enlace llamado "link_1" y se define su conexión con el "base_link" mediante una articulación:

Visual y Colisión: "link_1" es un cubo de dimensiones 0.05 x 0.05 x 0.5 metros. Está desplazado en x por 0.025 metros y en z por 0.25 metros. El material asignado es Cian con RGBA (0.318, 0.82, 0.965, 1.0).

Inercial: Masa de 0.3 kg y matriz de inercia similar al "base_link".

Articulación ("base_link_to_link_1"): Es de tipo revoluta, permitiendo rotación alrededor del eje definido por axis xyz="1 0 0" (eje x). Conecta "base_link" (padre) con "link_1" (hijo). La articulación está ubicada en (0.05, 0, 0.15).

```js
    <!--Definicion del Primer eslabon -->
    <link name="link_1">
        <!--Descripcion de la parte visual del eslabon-->
        <visual>
            <!--Posicion y orientación de origen del eslabon-->
            <origin xyz="0.025 0.0 0.25" rpy="0.0 0.0 0.0"/>
            <!--Definición de la geometria del eslabón-->
            <geometry>
                <box size="0.05 0.05 0.5"/>             
            </geometry>
            <!--Definicion del color del eslabon-->
            <material name="cian">
                <color rgba="0.318, 0.82, 0.965" />
            </material>      
        </visual>
        <!--Descripcion de las relaciones de colision-->
        <collision>
            <!--Posicion y orientacion de origen del eslabon-->
            <origin xyz="0.025 0.0 0.25" rpy="0.0 0.0 0.0"/>
            <!--Definicion de la geometria del eslabon-->
            <geometry>
                <box size="0.05 0.05 0.5"/>             
            </geometry>          
        </collision>
        <!--Descripcion de la inercial del eslabon-->
        <inertial>
            <!--masa del eslabon-->
            <mass value="0.3"/>
            <!--Ubicacion del sistema inercial asociado al centro de masa-->
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
            <!--Matriz de inercia en el centro de masa-->
            <inertia ixx="0.1" ixy="0.0" ixz="0.0" iyy="1.0" iyz="0.0" izz="1.0"/> 
        </inertial>
    </link>

   <joint name="base_link_to_link_1" type="revolute">
    <parent link="base_link"/>
    <child link="link_1"/>
    <axis xyz="1 0 0"/>
    <limit lower="-3.14159" upper="3.14159" velocity="50.0" effort="1000.0"/>
    <origin xyz="0.05 0 0.15" rpy="0 0 0"/>
   </joint>

```

4. La explicación sigue en la misma línea, detallando la estructura de "link_2", sus propiedades visuales, de colisión e inerciales, así como su conexión al primer enlace mediante una articulación de tipo revoluta.

Visual y Colisión: Similar al primer enlace (link_1), pero con un desplazamiento en x de 0.030 metros y en z de 0.25 metros. El material es de color rosa con valores RGBA (0.1, 0.753, 0.796, 1.0).

Inercial: La masa del segundo enlace es de 0.3 kg, con una matriz de inercia definida de manera similar al primer enlace.

Articulación ("link_1_to_link_2"): Esta articulación permite rotación alrededor del eje x, conectando "link_1" (como enlace padre) con "link_2" (como enlace hijo). Se establece un límite de rotación de -3.14159 a 3.14159 radianes y una velocidad máxima de 50.0 unidades, con un esfuerzo máximo de 1000 unidades.

```js
    <!--Definicion del Segundo Link-->
    <link name="link_2">
        <!--Descripcion de la parte visual del eslabon-->
        <visual>
            <!--Posicion y orientación de origen del eslabon-->
            <origin xyz="0.030 0.00 0.25" rpy="0.0 0.0 0.0"/>
            <!--Definición de la geometria del eslabón-->
            <geometry>
                <box size="0.05 0.05 0.5"/>             
            </geometry>
            <!--Definicion del color del eslabon-->
            <material name="blue">
                <color rgba="0.0 0.5 0.0 1.0" />
            </material>      
        </visual>
        <!--Descripcion de las relaciones de colision-->
        <collision>
            <!--Posicion y orientacion de origen del eslabon-->
            <origin xyz="0.030 0.05 0.25" rpy="0.0 0.0 0.0"/>
            <!--Definicion de la geometria del eslabon-->
            <geometry>
                <box size="0.05 0.05 0.5"/>             
            </geometry>          
        </collision>
        <!--Descripcion de la inercial del eslabon-->
        <inertial>
            <!--masa del eslabon-->
            <mass value="0.3"/>
            <!--Ubicacion del sistema inercial asociado al centro de masa-->
            <origin xyz="0.0 0.0 0.0" rpy="0.0 0.0 0.0"/>
            <!--Matriz de inercia en el centro de masa-->
            <inertia ixx="0.1" ixy="0.0" ixz="0.0" iyy="1.0" iyz="0.0" izz="1.0"/> 
        </inertial>
    </link>

   <joint name="link_1_to_link_2" type="revolute">
    <parent link="link_1"/>
    <child link="link_2"/>
    <axis xyz="1 0 0"/>
    <limit lower="-3.14159" upper="3.14159" velocity="50.0" effort="1000.0"/>
    <origin xyz="0.05 0 0.5" rpy="0 0 0"/>
   </joint>

```


5. se integran componentes para la simulación en Gazebo y se definen los sistemas de control del robot:

Definiciones de Gazebo: Se especifican materiales visuales en Gazebo para los enlaces "base_link" y "link_1". Por ejemplo, se asigna el material "Gazebo/Grey" al "base_link" y "Gazebo/Red" al "link_1".

Transmisión: Se define una transmisión simple llamada "trans_1". Está asociada a la articulación "base_link_to_link_1", y su interfaz de hardware es para control de posición (PositionJointInterface). Un actuador llamado "motor_1" se asigna a esta transmisión, con una reducción mecánica de 1.

Plugins de Gazebo: Dos plugins son añadidos. El primero es el "joint_state_publisher_controller", que publica los estados de las articulaciones con una frecuencia de 10 Hz. El segundo es el "joint_pose_trajectory_controller", que controla la trayectoria de las articulaciones y tiene una frecuencia de actualización de 2 Hz.

```js
    <!--Definicion en Gazebo-->
    <gazebo reference="base_link">
        <material>Gazebo/Grey</material>
    </gazebo>

    <gazebo reference="link_1">
        <material>Gazebo/Red</material>
    </gazebo>

    <!--Fin de la definicion de Gazebo-->

    <!--Definicion del control-->

    <transmission name ="trans_1">
        <type>transmission_interface/SimpleTransmission</type>
        <joint name="base_link_to_link_1">
            <hardwareInterface>hardware_interface/PositionJointInterface</hardwareInterface>
        </joint>
        <actuator name="motor_1">
            <hardwareInterface>hardware_interface/PositionJointInterface</hardwareInterface>
            <mechanicalReduction>1</mechanicalReduction>
        </actuator>
    </transmission>

    <gazebo>
        <plugin name="joint_state_publisher_controller"
                filename="libgazebo_ros_joint_state_publisher.so">
            <!-- Update rate in Hertz -->
            <update_rate>10</update_rate>
            <!-- Name of joints in the model whose states will be published. -->
            <joint_name>base_link_to_link_1</joint_name>
        </plugin>
    </gazebo>

    <gazebo>
        <plugin name="joint_pose_trajectory_controller"
                filename="libgazebo_ros_joint_pose_trajectory.so">
            <!-- Update rate in Hz -->
            <update_rate>2</update_rate>
        </plugin>
    </gazebo>

```
